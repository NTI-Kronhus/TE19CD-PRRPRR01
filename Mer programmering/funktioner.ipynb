```python
def medelvärde(a,b,c,d):
    medel = (a+b+c+d)/4 #lokal variabel, syns ej utanför funktionen
    return medel

#print(medel)
print(medelvärde(1,2,3,4))
print(medelvärde(10,20,30,40))
```

```python
# räkna ord 
def räkna_ord(text):
    ord = text.split()
    antal_ord = len(ord)
    return antal_ord

citat = "I stand on the shoulders of giants"
print(f"Antal ord i citatet: {räkna_ord(citat)}")


```

```python
import numpy as np 
import matplotlib.pyplot as plt 

def N(t):
    return 11/(1+3.4*np.exp(-0.03*t))

# a)
print(f"Antal människor år 1950: {N(0)} miljarder")

# b) övre gräns?
år = 500

t = np.arange(år)

plt.plot(t, N(t))
plt.ylabel("Antal människor i miljarder")
plt.xlabel("År efter 1950")
plt.title("Modell för population")
plt.grid()
plt.show()

print(f"Övre gräns: {N(5000)} miljarder")
```

```python
def minsta(tal1, tal2):
    if tal1 < tal2:
        return tal2
    else:
        return tal1

print(minsta(2,5))
```
