# 2. Beskriv koden detaljerat

```python
import matplotlib.pyplot as plt #importerar modul för att rita grafer

# skapat funktionen f(x) = x^2-3x+1
def f(x): 
    return x**2-3*x+1

x = [i for i in range(10)] # skapar en lista med talen 0,1,2,...,9
y = [f(i) for i in x] # använder varje tal i listan x som inparameter till funktionen f(x)

plt.plot(x,y) #ritar grafen med varje x-värden från listan x och y-värden från listan y
plt.xlabel("x") # namnger x-axeln för x
plt.ylabel("y") # namnger y-axeln för y
plt.show() # visar grafen

# övrig kommentar: grafen bör ha en lämplig titel också
```

# 8. Pengar till sedlar/mynt
## Nedbrytning av problemet: 
En funktion: 
* inparam: heltal
* skriva ut sedlar/mynt
* return: inget

```python
def växla(summa):
    kontanter = [1000, 500, 200, 100, 50, 20, 10, 5, 1] # sedlar o mynt
    sedlar = ["tusenlappar", "femhundralappar", "tvåhundralappar", "hundralappar", "femtiolappar", "tjugolappar", "tiokronor", "femkronor", "enkronor"] 
    
    for i in range(len(kontanter)): 
        antal = summa//kontanter[i] # heltalsdivision 
        summa %= kontanter[i] # ger resten vid heltalsdivision per sedel
        print(f"{antal} {sedlar[i]}", end=", ") # skriver ut antal per sedel

växla(3899)
```
